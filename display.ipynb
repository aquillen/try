```python
###### TEST CLASS ######
class particle():
    
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
```

```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
from mpl_toolkits.mplot3d import Axes3D

window = 50

### Prepare the data for triangulation
npoints = 1000                   # Number of stuck particles
data = np.zeros([npoints,3])  
parr_up = []
parr_down = []

xu,xd = [],[]
yu,yd = [],[]
zu,zd = [],[]

### Fill particle location arrays
for i in range(npoints):
    r = 20
    phi = np.random.rand()*6.28
    theta = np.random.rand()*3.14
    
    x = r*np.cos(theta)*np.sin(phi)
    y = r*np.sin(theta)*np.sin(phi)
    z = r*np.cos(phi)
    
    if z>=0:
        parr_up.append(particle(x,y,z))
        xu.append(x)
        yu.append(y)
        zu.append(z)
    
    else:
        parr_down.append(particle(x,y,z))
        xd.append(x)
        yd.append(y)
        zd.append(z)

triangu = mtri.Triangulation(xu, yu)
triangd = mtri.Triangulation(xd, yd)

### Prepare the data for plotting
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')

### Plot points
ax.plot_trisurf(xu,yu,zu, cmap='jet')
ax.plot_trisurf(xd,yd,zd, cmap='jet')
#ax.scatter(x,y,z, marker='.', s=10, c="black", alpha=0.5)
ax.view_init(elev=60, azim=-45)

### Format the graph
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim3d([-window/2, window/2])
ax.set_ylim3d([-window/2, window/2])
ax.set_zlim3d([-window/2, window/2])

plt.show()
```
