```python
###### TEST CLASS ######
class particle():
    
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
```

```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
from mpl_toolkits.mplot3d import Axes3D

window = 100

### Prepare the data for triangulation
npoints = 2000                   # Number of stuck particles
data = np.zeros([npoints,3])  
parr = []
x = []
y = []
z = []

### Fill particle location arrays
for i in range(npoints):
    r = 20
    theta = np.random.rand()*6.28
    phi = np.random.rand()*3.14
    parr.append(particle(r*np.cos(theta)*np.sin(phi),r*np.sin(theta)*np.sin(phi),r*np.cos(phi)))
    x.append(parr[i].x)
    y.append(parr[i].y)
    z.append(parr[i].z)


triang = mtri.Triangulation(x, y)

### Prepare the data for plotting
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')

### Plot points
#ax.plot_trisurf(triang, z, cmap='jet')
ax.scatter(x,y,z, marker='.', s=10, c="black", alpha=0.5)
ax.view_init(elev=60, azim=-45)

### Format the graph
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim3d([-window/2, window/2])
ax.set_ylim3d([-window/2, window/2])
ax.set_zlim3d([-window/2, window/2])

plt.show()
```
